The marketing team needs help on a new project. They are experiencing a decrease in the number of matches, and they are trying to find a way to understand what makes people interested into each other.

They decided to run a speed dating experiment with people who had to give Tinder lots of informations about themselves that could ultimately reflect on ther dating profile on the app.

Tinder then gathered the data from this experiment. Each row in the dataset represents one speed date between two people, and indicates wether each of them secretly agreed to go on a second date with the other person.

**Scope of this project** üñºÔ∏è

Data was gathered from participants in experimental speed dating events from 2002-2004.<br> During the events, the attendees would have a four minute "first date" with every other participant of the 
opposite sex.<br> 

At the end of their four minutes, participants were asked if they would like to see their date again. They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests.

The dataset also includes questionnaire data gathered from participants at different points in the process. These fields include: demographics, dating habits, self-perception across key attributes, beliefs on what others find valuable in a mate, and lifestyle information.
```python
tinder_df = df.copy()

# Dataset shape
tinder_df.shape

# Show the first 5 rows of the dataframe.
tinder_df.head()
```

```python
# Selecting data relating to unique participants, grouping by the 'iid' feature.
# Keep only the columns : gender, age, field, field_cp, wave, round, position, match, from, career, career_c
# Keep first for gender, age, field, field_cp, wave, round, position, from, career, career_c
# Keep average for match
tinder_iid_df = tinder_df.groupby('iid').agg({'age': 'first', 'gender': 'first', 'field': 'first', 'field_cd': 'first', 'wave': 'first', 'position': 'first', 'positin1':'first', 'match': 'mean', 'from': 'first'}).reset_index()
```

##  Correlations

```python
# Create a copy of the dataframe
tinder_binary_df = df.copy()

# Select main features
features = ['match', 'gender', 'age', 'field_cd', 'wave', 'position', 'samerace', 'order', 'int_corr', 'age_o', 'match_es', 'like', 'prob', 'met']

# Select only the main features
tinder_binary_df = tinder_binary_df[features]

# Drop NaN values
tinder_binary_df = tinder_binary_df.dropna()

# Drop non numeric columns
tinder_binary_df = tinder_binary_df.select_dtypes(include=[np.number])

# Display shape
print('The shape of the dataset is:', tinder_binary_df.shape)
print('The original shape of the dataset is:', df.shape)

# Calculate point biserial correlation between match and the rest
point_biserial_corr = {}
for column in tinder_binary_df.columns:
    if column != 'match':
        point_biserial_corr[column] = pointbiserialr(tinder_binary_df[column], tinder_binary_df['match'])[0]
        
# Sort the dictionary by the absolute value of the correlation
point_biserial_corr = dict(sorted(point_biserial_corr.items(), key=lambda item: abs(item[1]), reverse=True))

# Display the point biserial correlation
point_biserial_corr

# Plot the point biserial correlation
plt.figure(figsize=(10, 10))
sns.barplot(x=list(point_biserial_corr.keys()), y=list(point_biserial_corr.values()))
plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Point Biserial Correlation')
plt.title('Point Biserial Correlation between Match and Other Features')
plt.show()

# The point biserial correlation is a measure of the strength and direction of the association between a binary variable and a continuous variable.
```

```python
# Here we observe that the features 'like' and 'prob' have the highest point biserial correlation with the 'match' feature. Followed by match espectation. All three are similar.
# It shows that the perception is very important in the decision making process.
# Then the age of the partner, and the age of the participant are also important.
# As well as the interest correlation, and the order of the date.
# These are the main elements affecting the decision making process from these correlation results.
```

#   Conclusion

The analysis of participant behavior and preferences within the dating event reveals significant insights into the dynamics of attraction and matching patterns. Participants display a pronounced tendency to match with partners who have similar expectations regarding key attributes such as attractiveness, sincerity, intelligence, fun, ambition, and shared interests. This alignment in attribute valuation underscores the importance of perceived compatibility in the initial stages of attraction, suggesting that individuals seek partners who mirror their own priorities and values in a potential match.

The analysis reveals a strong correlation between participants' matching patterns and their fields of study or professional domains, highlighting the significance of shared experiences and intellectual compatibility in forming romantic connections. This suggests that successful matches often stem from deeper commonalities beyond mere physical or superficial attributes.

Moreover, the structure of the matching event significantly influences outcomes, with a higher likelihood of matches occurring in the middle of the event. This can be attributed to participants' increasing comfort and openness as the event progresses, allowing them to connect more freely with potential partners.

In summary, the findings emphasize the critical roles of shared values, background compatibility, and the social dynamics of dating events in fostering successful romantic matches. These insights provide a deeper understanding of the complex factors that underpin the initial stages of romantic attraction and connection formation.
